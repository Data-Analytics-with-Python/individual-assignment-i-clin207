## 3. Removing Constant Columns (14pts)

* Identify any columns with constant values (no variance; all rows have the same value).

```python
#Dropping duplicates from natadata_df
natadata_df.drop_duplicates(inplace=True)
print("Size of dataset after dropping duplicates (records,features):", natadata_df.shape)

# Identify constant columns with no variance
constant_cols = [col for col in natadata_df.columns if natadata_df[col].nunique() == 1]
print("Constant columns:", constant_cols)

# Drop constant columns
natadata_df.drop(columns=constant_cols, inplace=True)

# Confirm remaining columns
print("Size of remaining dataset (records,features):", natadata_df.shape)
```

## 4. Add a new column (15pts)

* Add a new column **Customer_Days** to store the total number of days each customer has been with the company. You may follow the steps below:
  * use `pd.to_datetime(df["Dt_Customer"])` to convert the string variable to a *date-time* variable

  * use `today = datetime.today()` to find the date of today

  * use `(today - df["Dt_Customer"]).dt.days` to find the difference between today and the onbard date.

```python
from datetime import datetime

# Convert 'Dt_Customer' to date-time variable
natadata_df['Dt_Customer'] = pd.to_datetime(natadata_df['Dt_Customer'],format='%d-%m-%Y', errors='coerce')

# Get today's date
today = pd.to_datetime('today')

#Calculate total days each customer has been with the company
natadata_df['Customer_Days'] = (today - natadata_df['Dt_Customer']).dt.days
natadata_df.head()
```

## 5. Categorical Features (15 pts)

* For `Education` and `Marital_Status`,

  * find how many customers fall into each category
  * create pie charts to show the percentage of each category; make sure to show the percentage numbers on the chart with 2 decimal places

* What do you findings suggest about the company's target demographic?

```python
import matplotlib.pyplot as plt
import numpy as np

# Count customers by Education and Marital_Status
edu_counts = natadata_df['Education'].value_counts()
marital_counts = natadata_df['Marital_Status'].value_counts()

print("Customer count by Education:")
print(edu_counts)
print("\nCustomer count by Marital Status:")
print(marital_counts)

# Pie chart for Education
plt.figure(figsize=(6,6))
plt.pie(edu_counts, labels=edu_counts.index, autopct='%1.2f%%', colors=plt.cm.Paired.colors)
plt.title("Customer Distribution by Education")
plt.show()

# Pie chart for Marital Status (code below are written by referencing Gen AI code to customize pie chart presentation to improve clarity and visualization)
fig, ax = plt.subplots(figsize=(10, 8))

# Create pie chart
wedges, texts = ax.pie(marital_counts, startangle=90, colors=plt.cm.Pastel1.colors, wedgeprops={'edgecolor': 'white'})

# Prepare positions for labels
angles = np.array([(wedge.theta2 + wedge.theta1) / 2 for wedge in wedges])
x = np.cos(np.deg2rad(angles))
y = np.sin(np.deg2rad(angles))

# Sort by y to spread out labels nicely
label_order = np.argsort(y)
x = x[label_order]
y = y[label_order]
labels = marital_counts.index[label_order]
values = marital_counts.values[label_order]

# Plot labels outside, evenly spaced vertically
for i, (xi, yi, label, val) in enumerate(zip(x, y, labels, values)):
    percentage = val / marital_counts.sum() * 100
    ha = "left" if xi > 0 else "right"

    # spread small labels vertically (using i offset)
    ax.annotate(f"{label}: {percentage:.2f}%", xy=(xi, yi), xytext=(1.4 * np.sign(xi), yi + (0.08 * (i - len(x)/2))), horizontalalignment=ha, arrowprops=dict(arrowstyle="-", connectionstyle=f"angle,angleA=0,angleB={(xi+yi)*90}"),fontsize=11, weight='bold')

plt.title("Customer Distribution by Marital Status", fontsize=14)
plt.tight_layout()
plt.show()
```
